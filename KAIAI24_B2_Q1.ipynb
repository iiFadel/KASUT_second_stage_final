```python
import pandas as pd
import numpy as np
```

```python
data_df = pd.read_csv('bikers_data.csv')
```

```python
data_df.head()
```

```python
data_df.corr()
```

```python
data_df.dropna(axis=0, inplace=True)
data_y = data_df['Number of bikers'] # target
data_x = data_df.drop(['Number of bikers'], axis=1) # input features
```

```python
data_x.head()
```

```python
data_y
```

```python
def normalize(df):
  return (df-df.mean())/df.std()

def mean_squared_error(y_actual, y_pred):
    return (1/len(y_actual)) * (np.sum((y_actual - y_pred) ** 2))
```

```python
data_x = normalize(data_x)
data_y = normalize(data_y)
```

```python
data_x = data_x.drop(['Date'], axis=1)
#data_x = data_x.drop(['holiday'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.2, random_state=1)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
train_mse = mean_squared_error(y_train, y_train_pred)
test_mse = mean_squared_error(y_test, y_test_pred)
```

```python
print(f"Training MSE: {train_mse}")
print(f"Testing MSE: {test_mse}")
```
